```python
# importing libraries
import numpy as np
import pandas as pd
```

```python
# reading the CSV files and storing in a DataFrame
movies=pd.read_csv('tmdb_5000_movies.csv')
credits=pd.read_csv('tmdb_5000_credits.csv')
```

```python
movies.head()
```

```python
credits.head()
```

```python
# merging both the CSV files into 'movies'
movies=movies.merge(credits,on='title')
```

```python
movies.head()
```

```python
# Important columns required for recommendation
# genres
# id
# keywords
# title
# overview
# cast
# crew

movies=movies[['genres','movie_id','keywords','title','overview','cast','crew']]
# taking only the important columns for the recommending system
# other columns like release year are not as important for the recommending system
```

```python
movies.info()
```

```python
movies.head()
```

```python
# checking NULL values in the DataFrame
movies.isnull().sum()
```

```python
# removing all the NULL values from the DataFrame
movies.dropna(inplace=True)
```

```python
# checking if duplicate values exist or not
movies.duplicated().sum()
```

```python
movies.iloc[0].genres
```

```python
# function for only taking the names from the whole list
import ast
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
# updating the genre column
movies['genres']=movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
# updating the keywords columns
movies['keywords']=movies['keywords'].apply(convert)
```

```python
def convert3(obj):
    L=[]
    counter=0
    for i in ast.literal_eval(obj):
        if counter!=3:
            L.append(i['name'])
            counter+=1
        else:
            break
    return L
```

```python
# updating the cast columns
movies['cast']=movies['cast'].apply(convert3)
```

```python
def fetch_director(obj):
    L=[]
    counter=0
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            L.append(i['name'])
            break
    return L
```

```python
# updating the crew columns
movies['crew']=movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

```python
# updating the overvew columns
movies['overview']=movies['overview'].apply(lambda x : x.split())
```

```python
movies.head()
```

```python
# removing spaces from all the columns which we've choosed for recommending system

movies['genres']=movies['genres'].apply(lambda x: [i.replace(' ','') for i in x])
movies['keywords']=movies['keywords'].apply(lambda x: [i.replace(' ','') for i in x])
movies['cast']=movies['cast'].apply(lambda x: [i.replace(' ','') for i in x])
movies['crew']=movies['crew'].apply(lambda x: [i.replace(' ','') for i in x])
```

```python
movies.head()
```

```python
# combining all columns in the new 'tag' column
movies['tags']=movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']
```

```python
# creating a new final DataFrame with only 3 columns which are important for recommending engine
new_df=movies[['movie_id','title','tags']]
```

```python
new_df.head()
```

```python
# converting the list to string
new_df = new_df.copy()
new_df['tags'] = new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df['tags'][0]
```

```python
# converting all the strings to lowerCase for faster execution
new_df['tags']=new_df['tags'].apply(lambda x: x.lower())
```

```python
new_df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000, stop_words='english')
```

```python
cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors= cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors[0]
```

```python
len(cv.get_feature_names_out())
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vectors)
```

```python
# Distance of first movie with all other movies
similarity[1]
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6]
```

```python
def recommend(movie):
    movie_index=new_df[new_df['title']==movie].index[0]
    distances=similarity[movie_index]
    movies_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])
    for i in movies_list[1:6]:
        print(new_df.iloc[i[0]].title) 
```

```python
recommend('The Avengers')
```
